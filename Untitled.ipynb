```python
import numpy as np
import matplotlib.pyplot as plt

# Load the track data from the .npy file
track_data = np.load('reInvent2019_track.npy')

# Assuming the track_data is a 2D array where each row represents a point on the track
# Extract x and y coordinates for plotting
x = track_data[:, 0]
y = track_data[:, 1]

# Plot the track
plt.plot(x, y, label='Track')
plt.scatter(x, y, color='red', label='Data Points')  # Scatter plot to highlight data points
plt.title('Track Plot')
plt.xlabel('X-coordinate')
plt.ylabel('Y-coordinate')
plt.legend()
plt.show()
```

```python
import numpy as np

# Load the track data from the .npy file
track_data = np.load('reInvent2019_track.npy')

# Assuming the track_data is a 2D array where each row represents a point on the track,
# with the first two columns as coordinates (x, y) and the next columns as values

# Extract coordinates and values for each point along the track
for i, data_point in enumerate(track_data):
    x, y = data_point[:2]  # Assuming the first two columns are x and y coordinates
    values = data_point[2:]  # Assuming the remaining columns are values associated with each point
    
    print(f"Point {i + 1} - Coordinates: ({x}, {y}), Values: {values}")
```
